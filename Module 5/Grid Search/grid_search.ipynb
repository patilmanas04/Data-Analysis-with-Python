```python
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
```

```python
df = pd.read_csv(file_path, header=0)
```

```python
df.head()
```

```python
# Grid Search
parameters = [{'alpha': [0.001, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]}]

RidgeModel = Ridge()
Grid = GridSearchCV(RidgeModel, parameters, cv=4)
Grid.fit(df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']], df.price)

Grid.best_estimator_
scores = Grid.cv_results_
scores['mean_test_score']
```

```python
# Grid Search with different combinations of parameters
parameters1 = [{
    'alpha': [0.001, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]
}]

RidgeModel1 = Ridge()
Grid1 = GridSearchCV(RidgeModel1, parameters1, cv=4)
Grid1.fit(df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']], df.price)

Grid1.best_estimator_
scores1 = Grid1.cv_results_

for param, mean_val in zip(scores1['params'], scores1['mean_test_score']):
    print(param, "R^2 on test data: ", mean_val)
```
