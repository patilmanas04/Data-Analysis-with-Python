```python
import pandas as pd
```

```python
file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style", "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type", "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(file_path, names=headers)
```

```python
df.head(5)
```

```python
df[['length', 'width', 'height']].head(5)
```

```python
# Data normalization using simple feature scaling
# Formula: X_new = X_old / X_max
df.length = df.length / df.length.max()
```

```python
df[['length', 'width', 'height']].head(5)
```

```python
# Data normalization using min-max
# Formula: X_new = (X_old - X_min) / (X_max - X_min)
df.width = (df.width - df.width.min()) / (df.width.max() - df.width.min())
```

```python
df[['length', 'width', 'height']].head(5)
```

```python
# Data normalization using z-score
# Formula: X_new = (X_old - X_mean) / X_std
df.height = (df.height - df.height.mean()) / df.height.std()
```
