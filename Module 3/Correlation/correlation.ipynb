```python
import pandas as pd
import seaborn as sns
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
```

```python
df = pd.read_csv(file_path, header=0)
```

```python
df.head(10)
```

```python
# correlation between engine size and price using the regplot
x = "engine-size"
y = "price"
sns.regplot(x=x, y=y, data=df)

# This is a postive linear relationship
```

```python
# correlation between highway mpg vs price using regplot
x = "highway-mpg"
y = "price"
sns.regplot(x=x, y=y, data=df)

# This is a negative linear relationship
```

```python
# correlation between peak rpm and price using regplot
x = "peak-rpm"
y = "price"
sns.regplot(x=x, y=y, data=df)

# This is a weak correlation
```
