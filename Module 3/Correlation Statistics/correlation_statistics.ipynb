```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
```

```python
df = pd.read_csv(file_path, header=0)
```

```python
df.head(10)
```

```python
# Pearson Correlation has two values: correlation coefficient and p-value

# About Correlation Coefficient
# if coefficient is close to +1 then it is a large positive correlation 
# if coefficient is close to -1 then it is a large negative correlation 
# if coefficient is 0 then there is no correlation

# About p-value
# if p-value < 0.001 then strong certainity of correlation
# if p-value < 0.05 and p-value > 0.001 then moderate certainity of correlation
# if p-value > 0.05 and p-value < 0.1 then weak certainity of correlation
# if p-value > 0.1 then there is no certainity of correlation

# pearsonr() method
pearson_coef, p_value = stats.pearsonr(df.horsepower, df.price)
print("Pearson Coefficient: ", pearson_coef)
print("p-value: ", p_value)

# This means that there is a strong positive correlation betweent he horsepower and price
```
