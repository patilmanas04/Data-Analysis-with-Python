```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
```

```python
df = pd.read_csv(file_path, header=0)
```

```python
df.head()
```

```python
lm = LinearRegression()
```

```python
# Simple Linear Model
# X: Predictor variable
# Y: Target variable
X = df[['highway-mpg']]
Y = df.price
lm.fit(X, Y)
```

```python
# We can output a prediction
Yhat = lm.predict(X)
Yhat[0:6]
```

```python
# Value of intercept
b_0 = lm.intercept_
b_0
```

```python
# Value of slope
b_1 = lm.coef_
b_1
```

```python
# So the simple linear regression equation will be: Yhat = b_0 + b_1*x
# Predicting price for a value of x
predicted_price = b_0 + b_1*28
print(predicted_price)
```
